```python
# -*- coding: utf-8 -*-
# ğŸ“Š Táº¡o PowerPoint: "PhÃ¢n loáº¡i Email Spam" (báº±ng Naive Bayes)
# --------------------------------------------------------------
# YÃªu cáº§u: pip install python-pptx matplotlib numpy

from pptx import Presentation
from pptx.util import Inches, Pt, Cm
from pptx.dml.color import RGBColor
from pptx.enum.text import PP_ALIGN
import matplotlib.pyplot as plt
import numpy as np
import os

# ğŸ“‚ Táº¡o thÆ° má»¥c xuáº¥t file
out_dir = "./"
os.makedirs(out_dir, exist_ok=True)

# ===== Táº O CÃC HÃŒNH MINH Há»ŒA =====

def save_pipeline(path):
    plt.figure(figsize=(8,3), facecolor='black')
    ax = plt.gca()
    ax.set_facecolor('black')
    ax.axis('off')
    boxes = ["Thu tháº­p dá»¯ liá»‡u", "Tiá»n xá»­ lÃ½", "TrÃ­ch chá»n\nÄ‘áº·c trÆ°ng",
             "Huáº¥n luyá»‡n\nmÃ´ hÃ¬nh", "PhÃ¢n loáº¡i", "ÄÃ¡nh giÃ¡"]
    xs = np.linspace(0.05,0.95,len(boxes))
    for i,(x,b) in enumerate(zip(xs, boxes)):
        ax.add_patch(plt.Rectangle((x-0.06,0.35),0.12,0.3,
                                   color='#1f77b4', ec='white', linewidth=1.2))
        ax.text(x,0.5,b,ha='center',va='center',color='white', fontsize=11)
        if i < len(boxes)-1:
            ax.arrow(x+0.06,0.5, xs[i+1]-x-0.12, 0,
                     head_width=0.03, head_length=0.02, fc='white', ec='white')
    plt.savefig(path, bbox_inches='tight', facecolor='black', dpi=150)
    plt.close()

def save_naive_bayes(path):
    plt.figure(figsize=(6,4), facecolor='black')
    ax = plt.gca(); ax.set_facecolor('black'); ax.axis('off')
    ax.add_patch(plt.Rectangle((0.05,0.4),0.25,0.35, color='#9467bd', ec='white'))
    ax.text(0.175,0.575,"Email\n(Chuá»—i vÄƒn báº£n)", ha='center', va='center',
            color='white', fontsize=11)
    features = ["tá»«: giáº£m", "tá»«: miá»…n phÃ­", "tá»«: click", "tá»«: kháº©n"]
    for i,f in enumerate(features):
        ax.add_patch(plt.Circle((0.55,0.8 - i*0.2), 0.07,
                                color='#ff7f0e', ec='white'))
        ax.text(0.55,0.8 - i*0.2, f, ha='center', va='center',
                color='white', fontsize=9)
        ax.arrow(0.3,0.58 - i*0.04, 0.2, 0.18 - i*0.18,
                 head_width=0.02, head_length=0.02, fc='white', ec='white')
    ax.add_patch(plt.Rectangle((0.82,0.65),0.14,0.12, color='#2ca02c', ec='white'))
    ax.text(0.89,0.71,"P(Spam)=0.84", ha='center', va='center', color='white', fontsize=10)
    ax.add_patch(plt.Rectangle((0.82,0.45),0.14,0.12, color='#d62728', ec='white'))
    ax.text(0.89,0.51,"P(Ham)=0.16", ha='center', va='center', color='white', fontsize=10)
    plt.savefig(path, bbox_inches='tight', facecolor='black', dpi=150)
    plt.close()

def save_laplace(path):
    plt.figure(figsize=(6,4), facecolor='black')
    ax = plt.gca(); ax.set_facecolor('black')
    words = ["tá»« a (0)", "tá»« b (5)", "tá»« má»›i (0)"]
    counts = [0,5,0]; smoothed = [c+1 for c in counts]
    x = np.arange(len(words))
    ax.bar(x-0.2, counts, width=0.35, label='Äáº¿m gá»‘c', color='#1f77b4')
    ax.bar(x+0.2, smoothed, width=0.35, label='Sau Laplace (Î±=1)', color='#ff7f0e')
    ax.set_xticks(x); ax.set_xticklabels(words, color='white')
    ax.legend(facecolor='black', edgecolor='white', fontsize=9)
    for t in ax.get_legend().get_texts(): t.set_color('white')
    plt.title("VÃ­ dá»¥ Laplace Smoothing", color='white')
    plt.savefig(path, bbox_inches='tight', facecolor='black', dpi=150)
    plt.close()

def save_confusion(path):
    plt.figure(figsize=(4,4), facecolor='black')
    ax = plt.gca(); ax.set_facecolor('black')
    cm = np.array([[90,4],[1,5]])
    im = ax.imshow(cm, cmap='Oranges')
    for (i,j),val in np.ndenumerate(cm):
        ax.text(j,i,f"{val}",ha='center',va='center',color='black',fontsize=14)
    ax.set_xticks([0,1]); ax.set_xticklabels(["Dá»± Ä‘oÃ¡n Ham","Dá»± Ä‘oÃ¡n Spam"], color='white')
    ax.set_yticks([0,1]); ax.set_yticklabels(["Thá»±c sá»± Ham","Thá»±c sá»± Spam"], color='white')
    plt.title("Ma tráº­n nháº§m láº«n (Confusion Matrix)", color='white')
    plt.savefig(path, bbox_inches='tight', facecolor='black', dpi=150)
    plt.close()

def save_architecture(path):
    plt.figure(figsize=(8,4), facecolor='black')
    ax = plt.gca(); ax.set_facecolor('black'); ax.axis('off')
    ax.add_patch(plt.Rectangle((0.03,0.4),0.18,0.25, color='#17becf', ec='white'))
    ax.text(0.12,0.53,"Client\n(ngÆ°á»i dÃ¹ng)", ha='center', va='center', color='white')
    ax.add_patch(plt.Rectangle((0.28,0.4),0.22,0.25, color='#9467bd', ec='white'))
    ax.text(0.39,0.53,"Mail Server\n(Exchange / SMTP)", ha='center', va='center', color='white')
    ax.add_patch(plt.Rectangle((0.58,0.55),0.18,0.2, color='#2ca02c', ec='white'))
    ax.text(0.67,0.65,"Bá»™ lá»c\n(NAIVE BAYES)", ha='center', va='center', color='white')
    ax.add_patch(plt.Rectangle((0.82,0.4),0.15,0.25, color='#ff7f0e', ec='white'))
    ax.text(0.895,0.53,"Kho dá»¯ liá»‡u\n(Logs / Dataset)", ha='center', va='center', color='white', fontsize=9)
    ax.arrow(0.2,0.52,0.06,0, head_width=0.02, head_length=0.02, fc='white', ec='white')
    ax.arrow(0.5,0.52,0.06,0, head_width=0.02, head_length=0.02, fc='white', ec='white')
    ax.arrow(0.76,0.6,0.04,-0.06, head_width=0.02, head_length=0.02, fc='white', ec='white')
    ax.arrow(0.76,0.48,0.04,0.02, head_width=0.02, head_length=0.02, fc='white', ec='white')
    plt.savefig(path, bbox_inches='tight', facecolor='black', dpi=150)
    plt.close()

# Xuáº¥t cÃ¡c áº£nh
img_paths = {
    "pipeline": os.path.join(out_dir, "pipeline.png"),
    "naive": os.path.join(out_dir, "naive_bayes.png"),
    "laplace": os.path.join(out_dir, "laplace.png"),
    "confusion": os.path.join(out_dir, "confusion.png"),
    "architecture": os.path.join(out_dir, "architecture.png"),
}

save_pipeline(img_paths["pipeline"])
save_naive_bayes(img_paths["naive"])
save_laplace(img_paths["laplace"])
save_confusion(img_paths["confusion"])
save_architecture(img_paths["architecture"])

# ===== Táº O FILE POWERPOINT =====
prs = Presentation()
blank = prs.slide_layouts[6]

def bg_black(slide): slide.background.fill.solid(); slide.background.fill.fore_color.rgb = RGBColor(0,0,0)

def add_title(title, sub=None):
    s = prs.slides.add_slide(blank); bg_black(s)
    tx = s.shapes.add_textbox(Cm(1),Cm(3),Cm(28),Cm(4)).text_frame
    p=tx.paragraphs[0]; p.alignment=PP_ALIGN.CENTER
    r=p.add_run(); r.text=title; r.font.size=Pt(36); r.font.bold=True; r.font.color.rgb=RGBColor(255,255,255)
    if sub:
        st=s.shapes.add_textbox(Cm(1),Cm(8),Cm(28),Cm(2)).text_frame
        sp=st.paragraphs[0]; sp.alignment=PP_ALIGN.CENTER
        sr=sp.add_run(); sr.text=sub; sr.font.size=Pt(14); sr.font.color.rgb=RGBColor(200,200,200)

add_title("PhÃ¢n loáº¡i Email Spam\nbáº±ng Naive Bayes", "á»¨ng dá»¥ng Machine Learning trong lá»c thÆ° rÃ¡c")

def slide(title, bullets, img=None):
    s=prs.slides.add_slide(blank); bg_black(s)
    t=s.shapes.add_textbox(Cm(1),Cm(1),Cm(28),Cm(2)).text_frame
    p=t.paragraphs[0]; p.text=title; p.font.size=Pt(20); p.font.bold=True; p.font.color.rgb=RGBColor(255,255,255)
    tb=s.shapes.add_textbox(Cm(1),Cm(3),Cm(12),Cm(10)).text_frame
    for i,b in enumerate(bullets):
        para=tb.paragraphs[0] if i==0 else tb.add_paragraph()
        para.text=b; para.font.size=Pt(14); para.font.color.rgb=RGBColor(230,230,230)
    if img: s.shapes.add_picture(img,Cm(14),Cm(3),Cm(12),Cm(7))

slide("Email Spam lÃ  gÃ¬ vÃ  vÃ¬ sao cáº§n lá»c?", [
    "CÃ¡c tin nháº¯n khÃ´ng mong muá»‘n, gá»­i hÃ ng loáº¡t vÃ  cÃ³ thá»ƒ Ä‘á»™c háº¡i.",
    "Háº­u quáº£: lá»«a Ä‘áº£o, mÃ£ Ä‘á»™c, lÃ£ng phÃ­ tÃ i nguyÃªn, tá»•n tháº¥t kinh táº¿.",
    "Má»¥c tiÃªu: báº£o vá»‡ ngÆ°á»i dÃ¹ng vÃ  cáº£i thiá»‡n hiá»‡u suáº¥t cÃ´ng viá»‡c."
])
slide("CÆ¡ sá»Ÿ lÃ½ thuyáº¿t: Naive Bayes", [
    "Dá»±a trÃªn Ä‘á»‹nh lÃ½ Bayes: P(A|B)=P(B|A)P(A)/P(B).",
    "Giáº£ Ä‘á»‹nh 'ngÃ¢y thÆ¡': cÃ¡c Ä‘áº·c trÆ°ng Ä‘iá»u kiá»‡n Ä‘á»™c láº­p.",
    "Æ¯u Ä‘iá»ƒm: nhanh, Ã­t tÃ i nguyÃªn, hiá»‡u quáº£ vá»›i vÄƒn báº£n."
])
slide("Quy trÃ¬nh phÃ¢n loáº¡i (Pipeline)", [
    "1) Thu tháº­p dá»¯ liá»‡u (SpamAssassin, Enron...).",
    "2) Tiá»n xá»­ lÃ½: tokenization, loáº¡i stopwords.",
    "3) TrÃ­ch chá»n Ä‘áº·c trÆ°ng: BoW / TF-IDF.",
    "4) Huáº¥n luyá»‡n & Ä‘Ã¡nh giÃ¡: Accuracy, Precision, Recall, F1."
], img_paths["pipeline"])
slide("Naive Bayes: CÃ¡ch hoáº¡t Ä‘á»™ng", [
    "Chuyá»ƒn email thÃ nh táº­p tá»« (features).",
    "TÃ­nh P(Tá»«|Spam) vÃ  P(Tá»«|Ham) rá»“i káº¿t há»£p.",
    "GÃ¡n nhÃ£n theo xÃ¡c suáº¥t háº­u nghiá»‡m lá»›n hÆ¡n."
], img_paths["naive"])
slide("Laplace Smoothing (Xá»­ lÃ½ tá»« má»›i)", [
    "Giáº£i quyáº¿t xÃ¡c suáº¥t báº±ng 0 khi tá»« chÆ°a xuáº¥t hiá»‡n.",
    "CÃ´ng thá»©c: P(w|C)=(count(w,C)+Î±)/(count(C)+Î±V).",
    "ThÆ°á»ng dÃ¹ng Î±=1 Ä‘á»ƒ lÃ m má»‹n (Laplace)."
], img_paths["laplace"])
slide("Káº¿t quáº£ mÃ´ hÃ¬nh (TÃ³m táº¯t)", [
    "Accuracy: ~95% (vÃ­ dá»¥ thá»±c nghiá»‡m).",
    "Precision/Recall/F1 cao cho táº­p thá»­.",
    "ChÃº Ã½: káº¿t quáº£ phá»¥ thuá»™c dá»¯ liá»‡u vÃ  tiá»n xá»­ lÃ½."
], img_paths["confusion"])
slide("Kiáº¿n trÃºc há»‡ thá»‘ng lá»c spam", [
    "Mail server â†’ Bá»™ lá»c Naive Bayes â†’ Kho dá»¯ liá»‡u.",
    "Latency tháº¥p, cáº­p nháº­t mÃ´ hÃ¬nh thÆ°á»ng xuyÃªn.",
    "Káº¿t há»£p SVM/Deep Learning cho ngá»¯ cáº£nh phá»©c táº¡p."
], img_paths["architecture"])
slide("TÃ³m táº¯t & Káº¿t luáº­n", [
    "Naive Bayes váº«n lÃ  ná»n táº£ng hiá»‡u quáº£ cho lá»c spam.",
    "Tiá»n xá»­ lÃ½ vÃ  trÃ­ch chá»n Ä‘áº·c trÆ°ng quyáº¿t Ä‘á»‹nh hiá»‡u quáº£.",
    "HÆ°á»›ng tiáº¿p theo: tÃ­ch há»£p mÃ´ hÃ¬nh ngá»¯ cáº£nh vÃ  phÃ²ng thá»§ AI táº¡o sinh."
])

# ğŸ’¾ LÆ°u file
out_path = os.path.join(out_dir, "PhÃ¢n loáº¡i Email Spam.pptx")
prs.save(out_path)
print(f"âœ… ÄÃ£ táº¡o file: {out_path}")
```
